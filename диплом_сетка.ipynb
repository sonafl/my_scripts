```python
got_data = pd.read_excel('сетка (7).xlsx', sheet_name='Лист3')

sources = got_data['Source']
targets = got_data['Target']
weights = got_data['Weight']

edge_data = zip(sources, targets, weights)
```

```python
 for e in edge_data:
    src = e[0]
    dst = e[1]
    w = e[2]

    got_net.add_node(src, src, title=src, size=400, color='#f2f2f2', borderWidthSelected=5 )
    got_net.add_node(dst, dst, title=dst, size=400, color='#f2f2f2', borderWidthSelected=5)
    #5620c3
    #cd00cd
    if w == 90:
        got_net.add_edge(src, dst, value=w*100, color='#5620c3', width=w/20)
    elif w == 50:
        got_net.add_edge(src, dst, value=w*100, color='#5620c3', width=w/20)
    elif w==20:
         got_net.add_edge(src, dst, value=w*100, color='#fc0fc0', width=w/50)
    
# got_net.show_buttons()

# https://pyvis.readthedocs.io/en/latest/documentation.html#pyvis.network.Network.get_adj_list
neighbor_map = got_net.get_adj_list()
```

```python
# как сделать описание:
# если сорс - то связанные профессии и секторы
# если таргет - (если сорс == 2), то в нескольких секторах и определение, если сорс = 1 то только определение
```

```python
for node in got_net.nodes:
    if node['title'] in sources.to_list():
        node['title'] += '\n\nСвязанные профессии и секторы: \n' + '\n'.join(neighbor_map[node['id']])
    elif node['title'] in targets.to_list() and len(got_data[got_data['Target']== node['id']]['Source'].unique())> 1:
        node['title'] += f"\n{definition[node['title']]}\n"+'\n\nСвязанные профессии и секторы: \n' + '\n'.join(neighbor_map[node['id']])
    else:
        node['title'] += f"\n{definition[node['title']]}\n"
    node['value'] = len(neighbor_map[node['id']])*2
#     print(node['title'])
#     print(node['value'])
```
